En exploitant les données j'ai remarqué que la Russie n'était pas présente dans le dataframe de données finale, dûe à une mauvaise normalisation du nom du pays dans ce dataframe

```python
# Ajouter une nouvelle catégorie
df_owid['country'] = df_owid['country'].cat.add_categories(['Russian Federation'])

# Remplacer les valeurs
df_owid.loc[df_owid['country'] == 'Russia', 'country'] = 'Russian Federation'
```

Pour fusionner avec les deux autres bases de données nous avons besoin du code pays, permettant ainsi d'identifier d'une unique façon les pays selon la norme ISO 3166, grâce au package pycountry.

```python
## pip install pycountry
import pycountry

# Fonction pour récupérer le code pays ISO 3166
def get_country_code(country_name):
    try:
        # Recherche dans pycountry
        country = pycountry.countries.lookup(country_name)
        return country.alpha_3  # Retourne le code à 3 lettres
    except LookupError:
        return None  # Si le pays n'est pas trouvé

# Ajouter une colonne 'code_pays' avec les codes ISO 3166
df_owid['code_pays'] = df_owid['country'].apply(get_country_code)

data_owid = df_owid
df_owid
```
