Cependant, l'access token n'est valide que temporairement, et mon compte étant gratuit, je n'ai pas accès à tout. En accord avec ce que la documentation de MFW recommande, nous allons donc à la place utiliser leur package (qui est directement dans Python), afin d'avoir accès à toutes les données qu'il nous faut. 

```python
results = catalog.find("tree")
data = results.iloc[0].load()
data = data.reset_index()
data
```

```python
df_owid=pd.DataFrame(data)
df_owid
```

Pour pouvoir faire la régression et lier ces données aux autres, nous allons remplacer les données vides par 0 

```python
# Remplacer toutes les valeurs NaN par 0
df_owid = df_owid.apply(lambda col: col.fillna(0) if col.isnull().any() else col)
df_owid
```

On exporte les données df_owid en csv

```python
# Export des données en csv
df_owid.to_csv('df_owid.csv', index=False)
```
