```python
# Importer les données obtenues précédement
data_BM = pd.read_csv('/home/onyxia/work/Deforestation/Données API/data_BM.csv')
data_FAO_pivot = pd.read_csv('/home/onyxia/work/Deforestation/CSV_FAO/data_FAO_pivot.csv', sep=";")



# On merge les données de la FAO et de la Banque Mondiale
data = pd.merge(data_BM, data_FAO_pivot, left_on=['code_pays', 'date'], right_on=['code_pays', 'Année'], how='inner')
data
```
