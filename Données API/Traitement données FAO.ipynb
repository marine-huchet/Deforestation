Nous allons ensuite devoir traiter les données. Pour y voir plus clair, nous allons créer un dataframe où une colonne correspond à un type de données (au lieu d'une seule colonne avec un vecteur). Pour cela, nous allons installer le package suivant :

Puis nous allons transformer "clair" en un dataframe plus lisible. 

```python
import pandas as pd
import ast
import re

# Fonction pour nettoyer les données
def clean_subdimensions(value):
    if isinstance(value, str):  # Si la valeur est une chaîne
        match = re.match(r"subdimensions\s+\{(.*)\}", value)
        if match:
            content = match.group(1)
            pairs = [item.split(":") for item in content.split(",")]
            return {k.strip(): v.strip() for k, v in pairs}
        else:
            # Essayer de convertir avec ast.literal_eval si applicable
            try:
                return ast.literal_eval(value)
            except (ValueError, SyntaxError):
                return None
    elif isinstance(value, dict):  # Si la valeur est déjà un dict
        return value
    return None  # Si le type n'est pas géré

# Appliquer la fonction de nettoyage
clair[2] = clair[2].apply(clean_subdimensions)

# Normaliser la colonne 2
subdimensions_df = pd.json_normalize(clair[2])

# Fusionner avec le DataFrame original
clair_df = pd.concat([clair.drop(columns=[2]), subdimensions_df], axis=1)

# Afficher le résultat
clair_df

```
