Nous allons récupérer les données de la Banque Mondiale de deux manières différentes : la première avec un seul indicateur (au choix) pour cibler, la seconde avec tous les indicateurs en même temps pour avoir une vue d'ensemble de ce qui est disponible. 

```python
import requests
import pandas as pd

# URL de l'API
url = "https://api.worldbank.org/v2/country/all/indicator/NY.GDP.MKTP.CD?date=1970:2020&format=json"

# Requête GET à l'API
req = requests.get(url)
data_eco = req.json()

# Vérification des données
if len(data_eco) > 1 and "indicator" in data_eco[1][0]:
    # Extraction des données utiles
    records = data_eco[1]
    
    # Transformation des données en DataFrame pandas
    df = pd.DataFrame(records)

    # Affichage de colonnes pertinentes
    table = df[["date", "value", "indicator", "country"]]
    print(table)
else:
    print("Les données retournées ne contiennent pas les informations attendues.")
```

```python
import requests
import pandas as pd

# Base URL de l'API
base_url = "https://api.worldbank.org/v2/country/all/indicator/"

# Liste des indicateurs
indicators = [
    "SI.POV.GINI",        # Indice de GINI
    "AG.LND.AGRI.ZS",     # Terres agricoles (% de la surface)
    "AG.LND.FRST.ZS",     # Surface forestière (% de la surface)
    "NY.GDP.MKTP.CD",     # PIB (en dollars US)
    "NY.GDP.MKTP.KD.GZ"   # Croissance du PIB (%)
]

# Paramètres de la requête
params = {
    "date": "1970:2024",  # Période
    "format": "json"      # Format de réponse
}

# Création d'un DataFrame pour contenir toutes les données
all_data = pd.DataFrame()

# Boucle pour récupérer les données pour chaque indicateur
for indicator in indicators:
    # Construction de l'URL pour chaque indicateur
    url = f"{base_url}{indicator}"
    
    # Requête GET à l'API
    response = requests.get(url, params=params)
    data = response.json()
    
    # Vérification des données reçues
    if len(data) > 1 and "indicator" in data[1][0]:
        # Extraction des données utiles
        records = data[1]
        df = pd.DataFrame(records)
        
        # Ajout des données dans le DataFrame principal
        all_data = pd.concat([all_data, df], ignore_index=True)
    else:
        print(f"Les données pour l'indicateur {indicator} n'ont pas été récupérées.")
        #Ceci permet de savoir si l'API ne nous donne pas les informations pour l'un des indicateurs demandés. 

# Sélection et affichage des colonnes pertinentes
if not all_data.empty:
    table = all_data[["date", "value", "indicator", "country"]]
    print(table)
else:
    print("Aucune donnée exploitable n'a été récupérée.")
#Ceci permet une visualisation rapide de ce que nous avons obtenu (avant d'utiliser réellement le data frame)
```

Restructuration du Data Frame pour avoir une colonne par indicateur : 

```python
# Étape 1 : Extraction des noms des indicateurs et des pays
all_data["indicator_name"] = all_data["indicator"].apply(lambda x: x.get("value", None) if isinstance(x, dict) else None)
all_data["country_name"] = all_data["country"].apply(lambda x: x.get("value", None) if isinstance(x, dict) else None)

# Étape 2 : Vérification et suppression des lignes avec des valeurs manquantes
print("Colonnes avec valeurs manquantes avant nettoyage :")
print(all_data[["indicator_name", "country_name", "date", "value"]].isnull().sum())

# Suppression des lignes invalides
all_data = all_data.dropna(subset=["indicator_name", "country_name", "date", "value"])

# Vérification après suppression
print("Colonnes avec valeurs manquantes après nettoyage :")
print(all_data[["indicator_name", "country_name", "date", "value"]].isnull().sum())

# Étape 3 : Conversion des types de colonnes
all_data["date"] = pd.to_numeric(all_data["date"], errors="coerce")  # Assure que 'date' est numérique
all_data["value"] = pd.to_numeric(all_data["value"], errors="coerce")  # Assure que 'value' est numérique

# Vérification finale des valeurs manquantes
print("Colonnes avec valeurs manquantes après conversion des types :")
print(all_data[["indicator_name", "country_name", "date", "value"]].isnull().sum())

# Étape 4 : Pivotement du DataFrame
pivoted_data = all_data.pivot_table(
    index=["country_name", "date"],   # Index par pays et date
    columns="indicator_name",        # Les noms des indicateurs deviennent des colonnes
    values="value",                  # Les valeurs des indicateurs
    aggfunc="first"                  # Utilise la première valeur si plusieurs existent
).reset_index()

# Nettoyage des colonnes
pivoted_data.columns.name = None  # Supprime le nom des colonnes
pivoted_data.columns = [col.replace(" ", "_") for col in pivoted_data.columns]  # Remplace les espaces par des underscores

# Affichage d'un aperçu des données transformées
print("Aperçu des données transformées :")
print(pivoted_data.head())

```
