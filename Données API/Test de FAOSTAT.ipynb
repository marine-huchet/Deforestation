Ce document montre que la FAO n'a pas de données importable pour le moment, et explique donc notre utilisation de ces données via un fichier CSV plus direct. Le package FAOSTAT est inutilisable pour le moment. 

Commençons par importer les packages nécessaire

```python
import pandas as pd
import faostat 
```

Voici l'ensemble des types de données que possède la FAO :

```python
df = faostat.list_datasets_df()
df
```

Nous remarquons alors (et grâce au site de la FAO) que les données qui nous intéresse pour la déforestation sont celles codé "FO" (ligne 54), soit "Forestry Production and Trade"

```python
clair = pd.DataFrame(faostat.list_pars(code="FO"))
clair
```

Nous voyons donc qu'il y a 4 éléments important à retirer pour le code FO : "area", "element", "item" et "year"

```python
faostat.get_items(code="FO")
```

```python
faostat.get_elements(code="FO")
```

```python
faostat.get_years(code="FO")
```

```python
faostat.get_par_df("FO","area")
```

Ainsi, nous voyons les charactéristiques qu'ils nous faut. Cependant : 

```python
mypars = {'area': "004",
              'element': "5516",
              'item': "1627",
              'year': "1963"}
data = faostat.get_data_df(code='FO', pars=mypars, strval=False)
data
```

Comme nous pouvons le remarquer, les données sont entièrement vide - et nous pouvons facilement vérifier sur le site de la FAO que ces données existent bien, mais sont simplement complétement inaccessibles via l'API (et donc cette extension) actuellement. Cela vaut quel que soit les paramètres demandés tant l'API a été laissé à l'abandon par la FAO. 

Ceci explique donc pourquoi nous n'allons pas utiliser l'API ni l'extension FAOSTAT, et à la place utiliser un fichier CSV que nous allons traiter directement.
